```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets, models, layers
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
x_train[0]
```

```python
y_train[0]
```

```python
plt.imshow(x_train[0])
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
x_train[0]
```

```python
x_train_flattened = x_train.reshape(-1, 28*28)
x_test_flattened = x_test.reshape(-1, 28*28)
```

```python
x_train_flattened.shape
```

```python
x_train_flattened[0]
```

```python
def plot_figure(x, y, model, index):
    plt.figure(figsize=(15,2))
    x_flatten = x.reshape(-1, 28*28)
    y_pred = model.predict(x_flatten)
    y_pred_values = [np.argmax(i) for i in y_pred]
    print(y_pred_values[index], '- predicted value')
    plt.imshow(x[index])
    print(y[index], '- real value')
```

# ann

```python
x_train.shape
```

```python
ann = models.Sequential([
    layers.Dense(10, input_shape=(784,), activation='softmax')
])

ann.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

ann.fit(x_train_flattened, y_train, epochs=5)
```

```python
ann.evaluate(x_test_flattened, y_test)
```

```python
y_pred_ann = ann.predict(x_test_flattened)
y_pred_ann_values = [np.argmax(i) for i in y_pred_ann]
y_pred_ann_values
```

```python
plot_figure(x_test, y_test, ann, 50)
```

```python
cm = tf.math.confusion_matrix(y_test, y_pred_ann_values)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

# cnn

```python
x_train = np.expand_dims(x_train, axis=-1)
x_train = np.repeat(x_train, 3, axis=-1)
print(x_train.shape)

x_test = np.expand_dims(x_test, axis=-1)
x_test = np.repeat(x_test, 3, axis=-1)
print(x_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
cnn = models.Sequential([
    
    #cnn
    layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu', input_shape=(28,28,3)),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    
    #dense
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation = 'softmax')
    
    
])
```

```python
cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
cnn.fit(x_train, y_train, epochs=5)
```

```python
cnn.evaluate(x_test, y_test)
```

```python
y_pred_cnn = cnn.predict(x_test)
y_pred_cnn_values = [np.argmax(i) for i in y_pred_cnn]
y_pred_cnn_values
```

```python
x_test = np.squeeze(x_test)
```

```python
x_test.shape
```

```python
def plot_figure_cnn(x, y, model, index):
    plt.figure(figsize=(15,2))
    y_pred = model.predict(x)
    y_pred_values = [np.argmax(i) for i in y_pred]
    print(y_pred_values[index], '- predicted value')
    plt.imshow(x[index])
    print(y[index], '- real value')
```

```python
plot_figure_cnn(x_test, y_test, cnn, 100)
```

```python
cm = tf.math.confusion_matrix(y_test, y_pred_cnn_values)
cm
```

```python
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Prediction')
plt.ylabel('Truth')
```
